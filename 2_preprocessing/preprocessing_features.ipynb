# Removing features

With the exploratory data analysis we found out that some features are not relevant to the prediction of HeartDiseases, and that others (like phisicalHealth) are too explicit indicators of HeartDiseases. In this project we are going to try to predict heart diseases with the minimum number of complex features possible.

```python
heart_attack_2020_cleaned = heart_attack_2020.drop(columns=['AlcoholDrinking','MentalHealth','PhysicalHealth'])
heart_attack_2020_cleaned.head()
```

Furthermore, we remove stroke from the features as it is too good of an indicator.

```python
heart_attack_2020_cleaned_stroke = heart_attack_2020_cleaned.drop(columns=['Stroke'])
heart_attack_2020_cleaned_stroke.head()
```
